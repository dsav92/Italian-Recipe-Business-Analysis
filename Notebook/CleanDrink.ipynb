```python
import pandas as pd
import csv
import re
import nltk
import time
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('Dataset/dizionario_Bevande.csv',sep='delimiter', header=None, encoding='utf-8', engine='python')
df_1 = pd.read_csv('Dataset/Nuovi/Clean/Recipes_ALL_Clean.csv',sep='delimiter', header=None, encoding='utf-8', engine='python')
```

```python
recipes = set()
bevande = set()
recipes_def = set()
```

```python
for elem in df[0]:
    bevande.add(elem)
len(bevande)
```

```python
for elem in df_1[0]:
    recipes.add(elem)
len(recipes)
```

```python
recipes_def = recipes.difference(bevande)
len(recipes_def)
```

```python
csvFilename = "Dataset/Nuovi/Clean/Recipes_ALL_Clean_NoDrink.csv"  # what name you want to save your csv as
csv = open(csvFilename, "w", encoding = "utf-8")  # create or open csv, "w" to write strings
csv.write('')
```

```python
for recipe in recipes_def:
    csv.write(recipe + "\n")
```
