```python
import pandas as pd
import csv
import re
```

```python
df = pd.read_csv('Dataset/Recipes_ALL.csv',sep='delimiter', header=None)
df_1 = pd.read_csv('Dataset/Recipes_donnamoderna.csv',sep='delimiter', header=None)
df_2 = pd.read_csv('Dataset/recipes_lospicchiodaglio.csv',sep='delimiter', header=None)
df_3 = pd.read_csv('Dataset/RecipesCFan_01.csv',sep='delimiter', header=None)
df_4 = pd.read_csv('Dataset/RecipesCFan_02.csv',sep='delimiter', header=None)
df_5 = pd.read_csv('Dataset/RecipesCFan_03.csv',sep='delimiter', header=None)
df_6 = pd.read_csv('Dataset/RecipesCucch_ (1).csv',sep='delimiter', header=None)
df_7 = pd.read_csv('Dataset/RecipesCucch_ (2).csv',sep='delimiter', header=None)
df_8 = pd.read_csv('Dataset/RecipesCucch_ (3).csv',sep='delimiter', header=None)
df_9 = pd.read_csv('Dataset/RecipesCucinaNat_ALL_1479.csv',sep='delimiter', header=None)
df_10 = pd.read_csv('Dataset/RecipesSalePepe_ALL_3504.csv',sep='delimiter', header=None)
```

```python
list_df = [df,df_1,df_2,df_3,df_4,df_5,df_6,df_7,df_8,df_9,df_10]
```

```python
#fornisce l'unione dei diversi csv creati attraverso l'estrazione dai siti, eliminando i duplicati

def getIntersect(ls):
    
    inter = set()
    
    for elem in ls[0][0]:
        inter.add(elem)
        
    for i in range(1,len(ls)):
        temp = set()
        
        for j in range(0,len(ls[i][0])):
            temp.add(ls[i][0][j])
            
        inter = inter.union(temp)
        
    return inter
```

```python
intersect = getIntersect(list_df)
```

```python
recipes = []

#pulisce il dataset eliminando caratteri speciali, punteggiatura, numerici e trasforma tutti i caratteri in minuscolo

for recipe in intersect:
    if(recipe != "recipes" and recipe != "Recipes"):
        temp = re.sub(u'[^a-zA-Z0-9àèùòìáéíóúÁÉÍÓÚâêîôÂÊÎÔãõÃÕçÇ: ]', '', recipe)
        temp = re.sub('[0-9]', '', temp)
        recipe = temp
        recipes.append(recipe.lower())
```

```python
print(len(recipes))
```

```python
# salva il nome delle ricette in un file csv

csvFilename = "Dataset/Recipes_ALL.csv"  
csv = open(csvFilename, "w")  
colNames = "Recipes\n"  
csv.write(colNames)
```

```python
for recipe in recipes:
    csv.write(recipe + "\n")
```
